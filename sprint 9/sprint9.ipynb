1. Calculadora de l'índex de massa corporal

Escriu una funció que calculi l'IMC ingressat per l'usuari/ària, és a dir, qui ho executi haurà d'ingressar aquestes dades. Pots obtenir més informació del seu càlcul en: 🔗 Índice de masa corporal IMC que es y como se calcula.

La funció ha de classificar el resultat en les seves respectives categories.

Consell: Intenta validar les dades prèviament, perquè enviï un missatge d’advertència si les dades introduïdes per l’usuari/a no estan en el format adequat o no pren valors raonables.

IMC=P/E2 E:estatura, p:peso 

Bajo peso: menos de 18,5.
Peso normal: 18,5-24,9.
Sobrepeso: 25-29,9.
Obesidad: más de 30.

Se calcula el índice de masa corporal (IMC) a partir del peso y la estatura.

Fórmula:
IMC = peso / estatura²

y se clasifica el resultado en diferentes categorías.


```python
import unicodedata

# Esta función pide los datos al usuario y valida que sean correctos
def obtener_datos():
    
    # Validación del peso
    while True:
        try:
            peso = float(input("Introduce tu peso en kg: "))
            
            # el peso es positivo?
            if peso > 0:
                break
            else:
                print("El peso debe ser un número positivo.")
        
        #  si el usuario introduce texto u otro formato incorrecto
        except:
            print("Introduce un número válido.")

    # Validación de la altura
    while True:
        try:
            altura = float(input("Introduce tu altura en metros: "))
            
            # altura positiva?
            if altura > 0:
                break
            else:
                print("La altura debe ser un número positivo.")
        
        #  si el usuario introduce un valor incorrecto error 
        except:
            print("Introduce un número válido.")

    
    return peso, altura


# Esta función calcula el IMC a partir del peso y la altura
def calcular_imc(peso, altura):
    
    # Fórmula del IMC: peso dividido entre altura al cuadrado
    imc = peso / (altura ** 2)
    
    # Redondeamos  resultado  2 decimales
    return round(imc, 2)


# clasificación del IMC en categorías
def clasificar_imc(imc):
    
    if imc < 18.5:
        return "Bajo peso"
    
    elif imc < 25:
        return "Normal"
    
    elif imc < 30:
        return "Sobrepeso"
    
    else:
        return "Obesidad"


# Función principal para todo el programa
def main():
    
    # Bucle para permitir repetir el cálculo varias veces
    while True:
        
        # Obtenemos  datos del usuario
        peso, altura = obtener_datos()
        
        # Calculamos el IMC
        imc = calcular_imc(peso, altura)
        
        # Clasificamos el resultado
        categoria = clasificar_imc(imc)

        # Mostramos el resultado final
        print(f"\nIMC: {imc} - Categoría: {categoria}\n")

        #quiere repetir?
        repetir = input("¿Quieres calcular otro IMC? (si/no): ")

        # Normalizamos el texto para evitar errores con acentos o mayúsculas
        repetir = unicodedata.normalize('NFKD', repetir)
        repetir = ''.join(c for c in repetir if not unicodedata.combining(c))
        repetir = repetir.lower()

        # Si el usuario no quiere continuar, salimos del programa
        if repetir != "si":
            print("Hasta la próxima")
            break


main()

```

2. Convertidor de temperatures

Existeixen diverses unitats de temperatura utilitzades en diferents contextos i regions. Les més comunes són Celsius (°C), Fahrenheit (°F) i Kelvin (K). També existeixen altres unitats com Rankine (°Ra) i Réaumur (°Re).

Selecciona almenys 2 conversors, de tal manera que en introduir una temperatura retorni, com a mínim, dues conversions, de manera que es puguin guardar (recorda que un print() no es pot guardar mai).

Consell: Intenta validar les dades prèviament, perquè enviï un missatge d’advertència si les dades introduides per l’usuari/a no estan en el format adequat.

(EXTRA): Pensa una manera d’emmagatzemar totes les possibles conversions en un sol objecte (Llista? Diccionari? DataFrame?) en comptes d’escriure molts if else en funció de la temperatura d’origen i la temperatura de destí.

## sistema de conversión entre Celsius, Fahrenheit y Kelvin

```python
# Esta función convierte una temperatura a otras unidades
def conversor(temp, unidad):

    # Validamos que la temperatura sea un número
    try:
        temp = float(temp)

    except:
        return "Error: introduce una temperatura válida."

    # Convertimos la unidad a mayúscula para evitar errores
    unidad = unidad.upper()

    # Diccionario para guardar resultados 
    resultado = {}

    # Conversión desde Celsius
    if unidad == "C":

        resultado["F"] = round((temp * 9/5) + 32, 2)
        resultado["K"] = round(temp + 273.15, 2)

    # Conversión desde Fahrenheit
    elif unidad == "F":

        # Primero convertimos a Celsius
        celsius = (temp - 32) * 5/9

        # Desp convertimos a otras unidades
        resultado["C"] = round(celsius, 2)
        resultado["K"] = round(celsius + 273.15, 2)

    # Si la unidad no existe mostramos error
    else:
        return "Error: unidad no válida."

    # Devolvemos el diccionario con las conversiones
    return resultado


# Ejecución del programa
print(conversor(30, "C"))
```

3. Comptador de paraules d'un text.

Escriu una funció que donat un text, mostri les vegades que apareix cada paraula. Intenta que es gestionin totes les casuístiques possibles que facin que el programa no funcioni correctament.

(EXTRA): Quina és la llargada mitjana de les paraules del text que has escrit? “Hola com va?” hauria de retornar (4+3+2) / 3 = 3

 se limpia el texto y se ceuntan las palabras, se calcula la longitud media de las palabaras en el extra


```python
import string


#  función que limpia el texto y elimina signos de puntuación
def limpiar_texto(texto):

    texto = texto.lower()

    signos = string.punctuation + "¿¡"

    for signo in signos:
        texto = texto.replace(signo, "")

    return texto


#  función cuenta cuántas veces aparece cada palabra
def contador_palabras(texto):

    diccionario = {}

    palabras = texto.split()

    for palabra in palabras:

        if palabra in diccionario:
            diccionario[palabra] += 1

        else:
            diccionario[palabra] = 1

    return diccionario


#  función calcula la longitud media de las palabras
def longitud_media(texto):

    palabras = texto.split()

    total_letras = 0

    for palabra in palabras:
        total_letras += len(palabra)

    media = total_letras / len(palabras)

    return round(media, 2)


#  función recibe y valida el texto del usuario
def recibir_texto():

    while True:

        texto = input("Introduce una frase: ").strip()

        # Validamos que el texto no esté vacío
        if texto == "":
            print("Error: introduce un texto válido.")

        else:
            return texto


# Función principal del programa
def main():

    texto = recibir_texto()

    texto_limpio = limpiar_texto(texto)

    conteo = contador_palabras(texto_limpio)

    media = longitud_media(texto_limpio)

    print("\nFrecuencia de palabras:\n")

    for palabra in conteo:
        print(f'"{palabra}" aparece {conteo[palabra]} veces')

    print(f"\nLongitud media de palabras: {media}")


# Ejecutamos el programa
main()
```

4. Diccionari invers (amb possibilitat de duplicats)

Resulta que el client té una enquesta molt antiga que s'emmagatzema en un diccionari i els resultats els necessita al revés, és a dir, intercanviant les claus i els valors. Els valors i claus en el diccionari original són únics; si aquest no és el cas, la funció hauria d'imprimir un missatge d'advertiment, juntament amb una llista amb els valors associats a la clau repetida.

Fixa’t en el següents exemples per veure l’output esperat.

Se invierten las claves y valores de un diccionario.
En caso de valores duplicados, se agrupan en listas.


```python
# Esta función invierte las claves y valores de un diccionario
def diccionario_inverso(diccionario_original):

    diccionario_nuevo = {}

    # Recorremos el diccionario original
    for clave in diccionario_original:

        valor = diccionario_original[clave]

        # Si el valor ya existe, significa que hay duplicados
        if valor in diccionario_nuevo:

            print(f"Advertencia: valor duplicado -> {valor}")

            # Añadimos la nueva clave a la lista
            diccionario_nuevo[valor].append(clave)

        else:
            # Creamos una lista con la primera clave
            diccionario_nuevo[valor] = [clave]

    # Convertimos listas de un solo elemento en valores simples
    for valor in diccionario_nuevo:

        if len(diccionario_nuevo[valor]) == 1:
            diccionario_nuevo[valor] = diccionario_nuevo[valor][0]

    # Devolvemos  diccionario invertido
    return diccionario_nuevo


# Diccionario de ejem
diccionario = {"a":1, "b":1, "c":3}

# Ejecutamos  función
resultado = diccionario_inverso(diccionario)

print("\nDiccionario invertido:")
print(resultado)
```

1. Comptador i endreçador de paraules d'un text.

El client va quedar content amb el comptador de paraules, però ara vol llegir arxius TXT i que calculi la freqüència de cada paraula ordenades dins de les entrades habituals del diccionari segons la lletra amb la qual comencen, és a dir, les claus han d'anar de la A a la Z i dins de la A hem d'anar de la A la Z. Per exemple, per a l'arxiu "tu_me_quieres_blanca.txt" la sortida esperada seria:

Se analiza un archivo de texto agrupando palabras por su letra inicial y ordenándolas.

```python
import string


# Esta función abre y lee el archivo de texto
def abrir_archivo(ruta):

    with open(ruta, "r", encoding="utf-8") as archivo:
        texto = archivo.read()

    return texto


# Esta función limpia el texto y elimina signos de puntuación
def limpiar_texto(texto):

    texto = texto.lower()

    signos = string.punctuation + "¿¡"

    for signo in signos:
        texto = texto.replace(signo, "")

    return texto


# Esta función agrupa y cuenta palabras según su letra inicial
def crear_diccionario(texto):

    palabras = texto.split()

    resultado = {}

    for palabra in palabras:

        letra = palabra[0]

        # Si la letra no existe, creamos el grupo
        if letra not in resultado:
            resultado[letra] = {}

        # Contamos palabras
        if palabra in resultado[letra]:
            resultado[letra][palabra] += 1

        else:
            resultado[letra][palabra] = 1

    return resultado


# Esta función ordena el diccionario alfabéticamente
def ordenar_diccionario(diccionario):

    resultado_ordenado = {}

    # Ordenar letras
    for letra in sorted(diccionario):

        resultado_ordenado[letra] = {}

        # Ordenar palabras
        for palabra in sorted(diccionario[letra]):

            resultado_ordenado[letra][palabra] = diccionario[letra][palabra]

    return resultado_ordenado


# Función principal
def main():

    texto = abrir_archivo("tu me quieres blanca.txt")

    texto_limpio = limpiar_texto(texto)

    diccionario = crear_diccionario(texto_limpio)

    resultado = ordenar_diccionario(diccionario)

    print(resultado)


# Ejecutamos el programa
main()

        
```

2. Conversió de tipus de dades.

El client rep una llista de dades i necessita generar dues llistes, la primera on estaran tots els elements que es van poder convertir en flotants i l'altra on estan els elements que no es van poder convertir. Exemple de la llista que rep el client:

Se separan los elementos convertibles a float de los que no lo son.

```python
lista = ['1.3', 'one', '1e10', 'seven', '3-1/2',
         ('2', 1, 1.4, 'not-a-number'),
         [1, 2, '3', '3.4']]


# Esta función separa los elementos convertibles a float de los que no se pueden convertir
def transformar_lista(lista):

    convertidos = []
    no_convertidos = []

    for elemento in lista:

        try:
            convertidos.append(float(elemento))

        except TypeError:

            for subelemento in elemento:

                try:
                    convertidos.append(float(subelemento))

                except:
                    no_convertidos.append(subelemento)

        except:
            no_convertidos.append(elemento)

    return convertidos, no_convertidos


resultado = transformar_lista(lista)

print(resultado)
```

1. Generador de Contrasenyes

Explora el funcionament del mòdul random de la llibreria numpy. 🔗 Random module in NumPy

En aquest punt, el client ha detectat un problema amb les contrasenyes que utilitzen els seus treballadors. Asdf1234, dates d’aniversari o similars. Per a solucionar-ho ens ha encarregat una funció de Python que generi contrasenyes més segures. La funció ha de dependre dels següents paràmetres:

Se genera una contraseña aleatoria configurable con diferentes caracteres.

```python
from numpy import random


# Esta función genera una contraseña aleatoria
def crear_contrasenya(longitud, mayusculas=True, minusculas=True, numeros=True, signos=False):

    # Validamos  la longitud correcta
    if longitud <= 0:
        return "Error: la longitud debe ser mayor que 0."

    # Conjuntos de caracteres
    mayus = list("ABCDEFGHIJKLMNOPQRSTUVWXYZ")
    minus = list("abcdefghijklmnopqrstuvwxyz")
    nums = list("0123456789")
    signos_list = list("!@#$%&*")

    pool = []

    # Lista para guardar la contraseña
    resultado = []

    # Añadimos caracteres según las opciones seleccionadas
    # y garantizamos al menos uno de cada tipo

    if mayusculas:
        pool += mayus
        resultado.append(random.choice(mayus))

    if minusculas:
        pool += minus
        resultado.append(random.choice(minus))

    if numeros:
        pool += nums
        resultado.append(random.choice(nums))

    if signos:
        pool += signos_list
        resultado.append(random.choice(signos_list))

    # Validamos que al menos haya un tipo de carácter
    if len(pool) == 0:
        return "Error: selecciona al menos un tipo de carácter."

    # Calculamos cuántos caracteres faltan
    restantes = longitud - len(resultado)

    # Completamos la contraseña con caracteres aleatorios
    for _ in range(restantes):

        resultado.append(random.choice(pool))

    # Mezclamos la contraseña para que no siga siempre el mismo orden
    resultado = random.permutation(resultado)

    # Convertimos la lista en un string
    contraseña = ""

    for caracter in resultado:
        contraseña += caracter

    # Devolvemos la contraseña final
    return contraseña


# Ejecutamos la función
print("contraseña:",crear_contrasenya(10, True, True, True, True))
```

2. Processament de dades simple

Una companya de feina ens ha demanat un favor, aprofitant que sap que estem aprenent a programar. Té un històric de partits de futbol català en un fitxer, on hi ha emmagatzemat els noms dels equips i els resultats. Necessita que processem les dades de manera automàtica, per extreure’n els resultats que necessita.

Utilitza el fitxer "historic_partits.txt"
Necessita un programa que retorni:

El nombre total de gols que ha fet cada equip.
El nom de l’equip més golejador.
El nom de l’equip més golejat
La classificació global (cada victòria: 3 pts, empat 1 pts, derrota 0 pts)



```python
# Esta función procesa los partidos del archivo
def procesar_partidos(ruta):

    # Diccionarios para guardar estadísticas
    goles = {}
    encajados = {}
    puntos = {}

    # Abrimos el archivo
    with open(ruta, "r", encoding="utf-8") as archivo:

        lineas = archivo.readlines()

    # Recorremos cada línea del archivo
    for linea in lineas:

        linea = linea.strip()

        # Saltamos líneas vacías
        if linea == "":
            continue

        partes = linea.split()

        # Buscamos dónde está el marcador
        for i in range(len(partes)):

            if "-" in partes[i]:

                resultado = partes[i]
                indice = i
                break

        # Construimos los nombres de los equipos
        equipo1 = " ".join(partes[:indice])
        equipo2 = " ".join(partes[indice + 1:])

        # Separamos los goles
        goles1 = int(resultado.split("-")[0])
        goles2 = int(resultado.split("-")[1])

        # Inicializamos equipos si no existen
        for equipo in [equipo1, equipo2]:

            if equipo not in goles:

                goles[equipo] = 0
                encajados[equipo] = 0
                puntos[equipo] = 0

        # Actualizamos goles marcados
        goles[equipo1] += goles1
        goles[equipo2] += goles2

        # Actualizamos goles recibidos
        encajados[equipo1] += goles2
        encajados[equipo2] += goles1

        # Calculamos puntos
        if goles1 > goles2:

            puntos[equipo1] += 3

        elif goles2 > goles1:

            puntos[equipo2] += 3

        else:

            puntos[equipo1] += 1
            puntos[equipo2] += 1

    # Equipo más goleador
    equipo_goleador = max(goles, key=goles.get)

    # Equipo más goleado
    equipo_goleado = max(encajados, key=encajados.get)

    # Clasificación ordenada por puntos
    clasificacion = sorted(puntos.items(), key=lambda x: x[1], reverse=True)

    # Devolvemos todos los resultados
    return goles, equipo_goleador, equipo_goleado, clasificacion


# Ejecutamos la función
resultado = procesar_partidos("historicpartits.txt")

print("\n--- Goles marcados ---")
print(resultado[0])

print("\nEquipo más goleador:")
print(resultado[1])

print("\nEquipo más goleado:")
print(resultado[2])

print("\n--- Clasificación ---")
print(resultado[3])
```
